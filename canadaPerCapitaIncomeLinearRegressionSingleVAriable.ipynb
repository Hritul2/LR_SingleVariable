- **Given year to per capita income of canada predict for next 10 years using LR in Single Variable**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('canada_per_capita_income.csv')
print(df)
print(df.columns)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
print(reg.coef_ )#slope of the line
print(reg.intercept_ )#intercept of the line
```

```python
%matplotlib inline
plt.ylabel('per capita income (US$)',fontsize=20)
plt.xlabel('year',fontsize=20)
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='*')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
```

```python
df_new= pd.read_csv('years.csv')
df_new
```

```python
prices=reg.predict(df_new)
print(prices)
```

```python
df_new['prices']=prices
df_new
```
